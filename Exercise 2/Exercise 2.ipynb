```python
import matplotlib.image as mpimg 
from matplotlib import pyplot as plt, colors
import numpy as np
%matplotlib inline
```

```python
def displayHsvImage(image):
    rgb = colors.hsv_to_rgb(image)
    plt.imshow(rgb)
```

```python
# a) reading image
img = mpimg.imread("oldtimer.png")
```

```python
plt.imshow(img)
```

```python
#b) Retaing only value
hsvImage = colors.rgb_to_hsv(img)
hsvImage[:,:,0:2] = 0
displayHsvImage(hsvImage)
```

```python
#c) Taking mean of RGB values and setting it as Value for an HSV image
grayScale = img.mean(axis=2)
grayScale = grayScale.reshape(grayScale.shape[0],grayScale.shape[1],1)
emptyArray = np.zeros((grayScale.shape[0],grayScale.shape[1],2))
ownGrayScaleHsv = np.concatenate((emptyArray,grayScale),axis=2)
displayHsvImage(ownGrayScaleHsv)
```

```python
#d) Reducing the Hue by 50 percent
hsvImage = colors.rgb_to_hsv(img)
hsvImage[:,:,1] /= 2
displayHsvImage(hsvImage)
```

```python
# color selected for blending
brown = np.zeros((img.shape[0],img.shape[1],3))
brown[:,:,0] = 230/255.0
brown[:,:,1] = 180/255.0
brown[:,:,2] = 130/255.0
plt.imshow(brown)
```

```python
# e) Multiplicative blending
rgbImage = colors.hsv_to_rgb(hsvImage)
newImage = rgbImage*brown
plt.imshow(newImage)
```

```python
#f) Blue and a Yello Image 
hsvBlue = colors.rgb_to_hsv(img)
hsvBlue[:,:,0] = hsvBlue[:,:,0]*0.6
displayHsvImage(hsvBlue)

plt.show()

hsvYellow = colors.rgb_to_hsv(img)
hsvYellow[:,:,0] = hsvYellow[:,:,0]*0.11
displayHsvImage(hsvYellow)

   
plt.show()

```
