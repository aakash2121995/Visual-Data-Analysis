```python
import pandas as pd

import seaborn as sns

from matplotlib import pyplot as plt

import numpy as np

import networkx as nx

%matplotlib inline
sns.set_style("darkgrid")
```

```python
df = pd.read_excel("breast-cancer-wisconsin.xlsx")
df.loc[df["bareNuc"].isnull(), "bareNuc"] = 1
```

## 4 a)

```python
corr = df.corr()

corr
```

## 4 b)

```python
## masking upper traingle to remove loops and redundant edges
mask = np.ones_like(corr, dtype=np.int)
mask[np.triu_indices_from(mask)] = False
lowerTraingle = mask*corr

## creating graph from the matrix
G=nx.from_pandas_adjacency(lowerTraingle)
correlationVals = nx.to_pandas_edgelist(G)["weight"]

## filtering edges based on correlation values
edgeFlags = 1*(correlationVals > 0.6)


## plotting figure
fig = plt.figure()
fig.set_figheight(20)
fig.set_figwidth(20)
nx.draw_spring(G, font_color='black',font_size=15, with_labels=True, node_size=300, node_color="skyblue", node_shape="s",
        alpha=0.9, linewidths=40,edge_color='w',width=edgeFlags)

fig.set_facecolor("#00000F")

plt.show()
```

## 4 c)

```python
width_val = edgeFlags*correlationVals

width_val = np.square(4*width_val)

fig = plt.figure()
fig.set_figheight(20)
fig.set_figwidth(20)
nx.draw_spring(G, font_color='black',font_size=12, with_labels=True, node_size=300, node_color="skyblue", node_shape="s",
        alpha=0.9, linewidths=40,width=width_val/2,edge_color=width_val,edge_cmap=plt.cm.winter)

fig.set_facecolor("#00000F")

plt.show()

```

## 4 d)

```python
corr = nx.to_pandas_adjacency(G)
maxCorr = corr.max()

highCorr = maxCorr > 0.9
mediumCorr = (maxCorr > 0.8) & (maxCorr <= 0.9)
lowCorr = (maxCorr > 0.6) & (maxCorr <=0.8)
ignoreCorr = maxCorr <=0.6

colorCoding = np.empty_like(maxCorr,dtype=int)
colorCoding[highCorr] = 3

colorCoding[mediumCorr] = 2
colorCoding[lowCorr] = 1
colorCoding[ignoreCorr] = 0
colorCoding


##plotting figure
fig = plt.figure()
fig.set_figheight(20)
fig.set_figwidth(20)
nx.draw_circular(G, font_color='black',font_size=20, with_labels=True, node_size=5000, node_color=colorCoding, node_shape="s",
        alpha=0.9, linewidths=40,width=width_val,edge_color=width_val,edge_cmap=plt.cm.winter,cmap=plt.cm.cool_r)

fig.set_facecolor("#00000F")

plt.show()

```

## 4 e)

### 1. Nodes 'Code' and 'Mitoses' are disconnected from the rest of the graph. It indicates that their correlation with rest of the attributes is less (less than the threshold).

### 2. No, if correlation is not transitive in nature. For example, as per the visualisation, 'class' is highly correlated with 'uniCels' and 'bareNuc', however 'uniCels' and 'bareNuc' are not strongly correlated

### 3. 'UniCels', 'bareNuc' and 'blaChroma'. Using 'UniCellShape' may not be a very good idea as it very strongly correlated with ''UniCels' and hence it would give a effect of using the same column twice, which may not help in generalisation of the model.
